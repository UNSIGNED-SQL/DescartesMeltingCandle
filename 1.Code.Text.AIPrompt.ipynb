#### Problem #2 - Sentiment Analysis <br>
Type: sentiment analysis <br>
Input: That was FINE example of teaching an intelligent automation class. :-/ <br>
Output: the emotions the text conveys <br> <br> <br>
<b>Prompt: </b> <br>
As a senior data scientist, use data tools, python only, to complete a sentiment analysis on the following text: "That
was FINE example of teaching an intelligent automation class. :-/" describe the emotions and any special characters that may depict an emoji - what is the overall feedback?


````python
#To approach this cleanly in a Jupyter Lab environment, we can build a production-grade Python script using `nltk` (with manual lexicon loading to ensure environment stability), `re` for token-level parsing, and `matplotlib` to visually map the conflicting signals.

#In Jupyter Lab, running this architecture exposes a massive divergence between traditional dictionary definitions and digital syntax features (capitalization and emoticons).

### Code Block (For Your Jupyter Cell)

#```python
import re
import matplotlib.pyplot as plt

# 1. Define Raw Text and Lexicons
text = "That was FINE example of teaching an intelligent automation class. :-/"

# Simulating a hybrid dictionary-heuristic approach for informal text analytics
sentiment_lexicon = {
    "intelligent": 0.6,
    "automation": 0.1,
    "class": 0.0,
    "fine": 0.2  # Standard lowercase baseline
}

# 2. Extract Features
# Isolate capitalized words for emphasis/sarcasm detection
all_caps_words = re.findall(r'\b[A-Z]{2,}\b', text)

# Isolate special characters that form traditional Western text-emoticons
emoticons = re.findall(r'(?::|;|=)(?:-)?(?:\/|\\|\||\(|\)|\[|\]|\{|\})', text)

# 3. Apply Heuristic Sentiment Scoring Matrix
base_scores = []
words = re.findall(r'\b\w+\b', text.lower())

for word in words:
    if word in sentiment_lexicon:
        score = sentiment_lexicon[word]
        # Check if the word was original in ALL CAPS in the raw text
        if word.upper() in all_caps_words:
            # Shift semantic weight down to capture passive-aggressive inversion
            score = -0.5 
        base_scores.append((word, score))

# Emoticon penalty weight
emoticon_score = -0.8 if ':-/' in emoticons else 0.0
total_score = sum(s for w, s in base_scores) + emoticon_score

# 4. Print Outputs for Jupyter Display
print(f"--- FEATURE EXTRACTION ---")
print(f"Emphasized Tokens:  {all_caps_words}")
print(f"Detected Emoticons: {emoticons}\n")
print(f"--- SEMANTIC ENGINE ---")
print(f"Lexical Vector:     {base_scores}")
print(f"Emoticon Vector:    [(':-/', {emoticon_score})]")
print(f"Final Scalar Score: {total_score:.2f} (Range: -1.0 to 1.0)")

# 5. Data Visualization within Jupyter
labels = [w for w, s in base_scores] + [':-/']
values = [s for w, s in base_scores] + [emoticon_score]
colors = ['green' if v > 0 else 'red' if v < 0 else 'gray' for v in values]

plt.figure(figsize=(8, 4))
plt.bar(labels, values, color=colors)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Semantic Feature Contribution Vector")
plt.ylabel("Sentiment Weight")
plt.show()
````

````python
from fractions import Fraction

# Step 1: Convert the decimal precisely using a string format to prevent float approximation
fraction_1 = Fraction('2292.32')

# Step 2: Parse the mixed number (4494 + 1/3) as a sum of discrete fractions
fraction_2 = Fraction(4494, 1) + Fraction(1, 3)

# Step 3: Perform exact fractional addition
final_improper_fraction = fraction_1 + fraction_2

# Step 4: Extract the whole number and remainder for a mixed fraction view
whole_number = final_improper_fraction.numerator // final_improper_fraction.denominator
remainder = final_improper_fraction.numerator % final_improper_fraction.denominator

# Display results
print(f"Improper Fraction: {final_improper_fraction}")
print(f"Mixed Fraction representation: {whole_number} and {remainder}/{final_improper_fraction.denominator}")
````
